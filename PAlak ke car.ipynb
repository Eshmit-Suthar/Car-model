```python
df['insurance_validity'] = df['insurance_validity'].map({'Comprehensive':0, 'Third party insurance':1, 'ZeroDep':2,'Not Available':3,'Third Party':1})
```

```python
df['insurance_validity'].value_counts()
```

```python
df['fuel_type']=df['fuel_type'].map({'Petrol':0, 'Diesel':1, 'CNG' :2})
```

```python
df['fuel_type'].value_counts()
```

```python
df['transmission']=df['transmission'].map({'Manual':0, 'Automatic':1})
```

```python
df['transmission'].value_counts()
```

```python
df['registration_year'] = df['registration_year'].str[-2:].astype(int)
```

```python
df['registration_year'].value_counts()
```

```python
df.head()
```

```python
df['ownsership'] = df['ownsership'].map({'First Owner':1, 'Second Owner':2, 'Third Owner':3, 'Fifth Owner':5 })
df['ownsership'].value_counts()
```

```python
X = df [['insurance_validity', 'fuel_type', 'kms_driven','ownsership','transmission']]
y=df['price(in lakhs)']
```

```python
X.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR

# Create an imputer to handle NaN values
imputer = SimpleImputer(strategy='mean')  # You can choose 'mean', 'median', 'most_frequent', or 'constant'

# Apply the imputer to X
X_imputed = imputer.fit_transform(X)

# Now create your models
models = {'LinReg': LinearRegression(),
         'KNN-3': KNeighborsRegressor(n_neighbors=3),
         'SVM-Poly': SVR(kernel='poly')}

# Train and evaluate each model
for name, model in models.items():
    model.fit(X_imputed, y)  # Use the imputed data
    # Note: 'sc' was not defined in your original code
    # Assuming you meant to calculate a score
    score = model.score(X_imputed, y)
    print(name, score)
```
